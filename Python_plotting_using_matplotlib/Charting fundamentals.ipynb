```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('resources/course2/assets/NYC hourly traffic.csv')

df['Date']=df['Date'].apply(pd.to_datetime)
df.head()
```

```python
sample=df.query("`Plaza ID`==5 & Date>'2016-12-30' & Date<'2017-05-01'")
sample

```

```python
mask = (df['Plaza ID'] == 5) & (df['Date'] > '2015-12-30') & (df['Date'] < '2017-01-01')
sample2 = df[mask]
sample2.head()
```

```python
plt.hist(sample['Hour'], bins=24,weights=sample["# Vehicles - E-ZPass"]);
```

```python
sample['Day of Week']=sample['Date'].dt.dayofweek

plt.hist(sample['Day of Week'],bins=7,weights=sample["# Vehicles - E-ZPass"]);
```

```python
plt.figure(figsize=(12,8))
plt.hist2d(sample['Hour'],
          sample['Day of Week'],
          bins=[24,7],
          weights=sample["# Vehicles - E-ZPass"])
plt.colorbar();
```

```python
sample2['Month'] = sample2['Date'].dt.month

plt.hist(sample2['Month'], bins=12, weights=sample2['# Vehicles - E-ZPass'], align='left', rwidth=0.8);
```

```python
def week_of_month(dt):
    first_day = dt.replace(day=1)
    dom = dt.day
    adjusted_dom = dom + first_day.weekday()
    return int((adjusted_dom - 1) / 7) + 1

sample2['Week of Month'] = sample2['Date'].apply(week_of_month)

plt.hist(sample2['Week of Month'], bins=5, weights=sample2['# Vehicles - E-ZPass'], align='left', rwidth=0.8);
```

```python
plt.figure(figsize=(12,8))
plt.hist2d(sample2['Month'],
          sample2['Week of Month'],
          bins=[12,5],
          weights=sample2["# Vehicles - E-ZPass"])
plt.colorbar();
```

```python
from ipywidgets import interact
import ipywidgets as widgets

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('resources/course2/assets/wipeout.csv')
df=df.set_index(pd.to_datetime(df['timestamp']))
df.head()
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
potential_columns=[]
for col in df.columns:
    if df[col].dtype != 'obcject':
        potential_columns.append(col)
```

```python
def plot_data(col_names):
    plt.figure(figsize=[12,8])
    plt.gcf().suptitle(str(col_names))
    if len(col_names)==2:
        plt.plot(df[list(col_names)[0]],'b')
        ax=plt.twinx()
        ax.plot(df[list(col_names)[1]],'r')
    else:
        plt.plot(df[list(col_names)])
```

```python
plot_data(['cadence','enhanced_speed'])
```

```python
@interact(col_names=widgets.SelectMultiple(
    options=potential_columns,
    value=(potential_columns[0],),
    description='Columns'))
def plot_data(col_names):
    plt.figure(figsize=[12,8])
    plt.gcf().suptitle(str(col_names))
    if len(col_names) ==2:
        plt.plot(df[list(col_names)[0]],'b')
        ax=plt.twinx()
        ax.plot(df[list(col_names)[1]],'r')
    else:
        plt.plot(df[list(col_names)])
```
